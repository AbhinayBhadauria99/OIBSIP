```python
import speech_recognition as sr
import pyttsx3
import datetime
import webbrowser
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Initialize text-to-speech engine
def speak(text):
    """Convert text to speech."""
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

# Listen for user voice input
def listen():
    """Listen for user voice input."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        try:
            audio = recognizer.listen(source)
            command = recognizer.recognize_google(audio)
            print(f"You said: {command}")
            return command.lower()
        except sr.UnknownValueError:
            print("Sorry, I didn't catch that.")
            return ""
        except sr.RequestError:
            print("Network error.")
            return ""

# Process and execute user commands
def execute_command(command):
    """Process and execute user commands."""
    
    # Greet the user
    if "hello" in command or "hi" in command:
        speak("Hello! How can I assist you today?")
    
    # Normal conversation responses
    elif "how are you" in command:
        speak("I'm just a program, but I'm doing fine. How are you?")
    
    elif "what's up" in command or "what is up" in command:
        speak("Not much, just here to assist you!")
    
    elif "how are you doing" in command:
        speak("I'm doing well, thank you for asking!")
    
    # Respond to time command
    elif "time" in command:
        now = datetime.datetime.now().strftime("%H:%M:%S")
        speak(f"The current time is {now}")
    
    # Respond to date command
    elif "date" in command:
        today = datetime.date.today().strftime("%B %d, %Y")
        speak(f"Today's date is {today}")
    
    # Perform a web search
    elif "search" in command:
        speak("What should I search for?")
        query = listen()
        if query:
            webbrowser.open(f"https://www.google.com/search?q={query}")
            speak(f"Searching Google for {query}")
    
    # Email functionality
    elif "send email" in command:
        speak("What should I say in the email?")
        email_content = listen()
        
        # Always send email to abcd@gmail.com
        recipient = "abhinaybhadauria99@gmail.com"
        
        # Send email
        try:
            send_email(recipient, email_content)
            speak("Your email has been sent successfully!")
        except Exception as e:
            speak(f"Sorry, there was an error sending the email: {e}")

    # Exit the assistant
    elif "exit" in command or "bye" in command:
        speak("Let me know if anything else I can hep you with. Goodbye!")
        return False
    
    else:
        speak("I didn't understand that. Please try again.")
    
    return True

# Send email function
def send_email(recipient, content):
    """Send email to the specified recipient."""
    sender_email = "mroverledge@gmail.com"  # Change this to your email address
    sender_password = "pckvmxrumideapbr"  # Use App Password here if using 2FA for Gmail
    
    # Create email
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = recipient
    msg['Subject'] = "Message from your Voice Assistant"
    msg.attach(MIMEText(content, 'plain'))
    
    # Connect to Gmail SMTP server and send email
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()  # Secure the connection
        server.login(sender_email, sender_password)
        text = msg.as_string()
        server.sendmail(sender_email, recipient, text)
        server.quit()
    except Exception as e:
        raise Exception(f"Error sending email: {e}")

# Run the voice assistant
def start_assistant():
    """Run the voice assistant."""
    speak("Starting your voice assistant.")
    while True:
        command = listen()
        if not execute_command(command):
            break

if __name__ == "__main__":
    start_assistant()
```
